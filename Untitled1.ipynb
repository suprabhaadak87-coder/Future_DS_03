```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk
nltk.download('punkt')
```

```python
df = pd.read_csv('college_event_feedback_dataset.csv')
df.head()
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def get_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['polarity'] = df['Feedback'].apply(get_polarity)
df['subjectivity'] = df['Feedback'].apply(get_subjectivity)
df['Sentiment'] = df['polarity'].apply(get_sentiment)
df.head()
```

```python
plt.figure(figsize=(6,6))
df['Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen','lightcoral','lightblue'])
plt.title('Sentiment Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['Sentiment'].value_counts().plot(kind='bar', color=['lightgreen','lightcoral','lightblue'])
plt.title('Sentiment Count')
plt.xlabel('Sentiment')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.show()
```

```python
print("Total Feedback:", len(df))
print(df['Sentiment'].value_counts())
print("\nSentiment Percentage Distribution:")
print(round(df['Sentiment'].value_counts(normalize=True)*100,2))
```

```python
print("\nMost Positive Feedback:\n", df.loc[df['polarity'].idxmax(), 'Feedback'])
print("\nMost Negative Feedback:\n", df.loc[df['polarity'].idxmin(), 'Feedback'])
```
